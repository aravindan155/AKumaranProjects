```python
import random
import time

def drive_through_simulation():
    # Customer pulls up to the speaker
    print("*Customer vehicle approaches drive-through*")
    print("*Sensor detects vehicle*")
    print("*Notification sound plays in restaurant*")
    
    # Short delay before employee responds
    time.sleep(1.5)  # Simulates 1.5 second delay
    
    # Employee greets customer
    print("Welcome to Dosa Den, may I have your order?")
    
    # Call the order function
    order()

def choice():
    """Randomly select items from each menu category"""
    combos = ["Dosa combo", "Appam combo", "Puri combo", "Chapati combo", "Naan combo", "Rice combo"]
    sides = ["Yogurt rice", "Rasam", "Meat Sampler", "Veg Sampler"]
    drinks = ["Lassi", "Falooda", "Thumbs up", "Coke", "Fanta", "Sprite", "Maaza"]
    sweets = ["Rasgulla", "Jilebi", "Cashew sweet", "Ladu"]
    
    selected_combo = random.choice(combos)
    selected_side = random.choice(sides)
    selected_drink = random.choice(drinks)
    selected_sweet = random.choice(sweets)
    
    return selected_combo, selected_side, selected_drink, selected_sweet

def order():
    """Process a customer order with random selections"""
    # Get random menu selections
    combo, side, drink, sweet = choice()
    
    # Simulate customer speaking
    print("\nCustomer: Can I have a", combo)
    time.sleep(0.8)
    
    # Random chance to add additional items
    if random.random() > 0.3:
        print("Customer: With", side)
        time.sleep(0.8)
    
    if random.random() > 0.4:
        print("Customer: And a", drink)
        time.sleep(0.8)
    
    if random.random() > 0.7:
        print("Customer: Can I also get a", sweet)
        time.sleep(0.8)
    
    # Employee confirms order
    print("\nEmployee: Got it!.")
    time.sleep(1)
    
    # List all ordered items
    print(f"Employee: That's a {combo}")
    if random.random() > 0.3:
        print(f"Employee: With {side}")
    if random.random() > 0.4:
        print(f"Employee: A {drink}")
    if random.random() > 0.7:
        print(f"Employee: And a {sweet}")
    
    # Total price (random between $15-30)
    total = round(random.uniform(15, 30), 2)
    print(f"\nEmployee: Your total is ${10}. At the next window.")

# Run the simulation
drive_through_simulation()
```
