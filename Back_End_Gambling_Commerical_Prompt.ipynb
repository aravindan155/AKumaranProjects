```python
import random
import time

# Possible scenarios that can occur during the game
SCENARIOS = [
    "The dice roll off the table!",
    "One die lands on its edge...",
    "The dice glow mysteriously for a moment.",
    "You hear distant thunder as you shake the dice.",
    "The dice feel unusually warm in your hand.",
    "A sudden breeze passes through the room.",
    "The dice make an unusual sound when they land.",
    "Time seems to slow down as the dice roll.",
    "The dice bounce higher than expected.",
    "You feel a strange confidence about this roll."
]

def roll_dice(num_dice=6):
    """Roll multiple dice and return the results."""
    return [random.randint(1, 6) for _ in range(num_dice)]

def check_for_pair_of_fours(dice_values):
    """Check if there's at least a pair of 4s in the dice roll."""
    return dice_values.count(4) >= 2

def check_for_all_fours(dice_values):
    """Check if all dice show 4 (the mythical 4x4 luck pair)."""
    return all(value == 4 for value in dice_values)

def random_scenario():
    """Return a random scenario from the list."""
    return random.choice(SCENARIOS)

def display_roll(dice_values, roll_number, wager):
    """Display the dice roll in a readable format."""
    print(f"\nRoll #{roll_number}:")
    
    # Generate a random scenario
    if random.random() < 0.7:  # 70% chance of scenario
        print(random_scenario())
    
    print("Dice values:", dice_values)
    
    if roll_number == 4:
        # On the 4th roll, ensure there's a pair of 4s
        print("Winner !!!")
        print(f"You've won your initial wager of ${wager}!")
        print("\nDo you want to try for the mythical 4x4 luck pair and double your money?")
        print("(Y/N): Y")  # Auto-accepting for simulation purposes
        return True
    elif roll_number == 5:
        # The risky fifth roll - checking for the mythical 4x4 (all dice showing 4)
        if check_for_all_fours(dice_values):
            print(f"INCREDIBLE! You hit the mythical 4x4 luck pair!")
            print(f"Your wager doubles to ${wager * 2}!")
        else:
            print("You Lose...")
            print("Have a gambling problem...GET HELP")
            print("National Problem Gambling Helpline: 1-800-522-4700")
    elif check_for_pair_of_fours(dice_values):
        print("You got a pair of 4s, but not on the winning roll yet.")
    else:
        print("No pair of 4s yet.")
    
    return False

def main():
    # Ask for wager amount
    print("Welcome to the Dice Gambling Game!")
    wager = random.randint(50, 500)  # Random wager for simulation
    print(f"Your wager: ${wager}")
    
    max_rolls = 4  # Initially 4 rolls
    
    print("Starting the dice game!")
    print("4x4")
    print("4x4")
    print("4x4")
    print("4x4")
    
    # The execution code
    print("\nExecution code: Big Money!!!")
    
    for roll_number in range(1, max_rolls + 1):
        print(f"\nPreparing roll #{roll_number}...")
        time.sleep(1)  # Add a small delay for suspense
        
        if roll_number == 4:
            # On the 4th roll, ensure there's at least a pair of 4s
            dice_values = roll_dice(6)
            # Make sure at least two dice show 4
            if dice_values.count(4) < 2:
                # Replace the first two dice with 4s
                dice_values[0] = 4
                dice_values[1] = 4
        else:
            # Regular random roll for other attempts
            dice_values = roll_dice(6)
        
        go_for_double = display_roll(dice_values, roll_number, wager)
        
        # If we completed the 4th roll and won, offer the 5th roll
        if roll_number == 4 and go_for_double:
            print("\nRolling for double or nothing...")
            time.sleep(2)  # Dramatic pause
            
            # 5th roll is always rigged to fail (extremely unlikely to get all 4s)
            fifth_roll = roll_dice(6)
            display_roll(fifth_roll, 5, wager)
            break

if __name__ == "__main__":
    main()
```
