```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from  scipy import stats
#India Manufacturing Output Historical Data
India_Manufacturing_Historical_Output=pd.read_csv(r'C:\Users\Aru\AppData\Local\Temp\378824c3-48a7-4da7-9080-eb717c6bcf1c_archive (1).zip.f1c\10 - India_Manufacturing_Output_Historical_Data.csv')
#India Historical Economic Growth
India_Historical_Economic_Growth=pd.read_csv(r'C:\Users\Aru\AppData\Local\Temp\bed72c11-a357-4b55-a747-872f55248bb0_archive (1).zip.bb0\1 - India_Economic_Growth_Historical_Data.csv')
#Merge
India = India_Historical_Economic_Growth.merge(
    India_Manufacturing_Historical_Output[['year', 'billions_of_us']],  # Only select year and billions_of_us from df2
    on='year',
    how='inner'  # Only keep years that exist in both DataFrames
)[['year', 'GDP', 'billions_of_us']]
stats.spearmanr(India['GDP'], India['billions_of_us'])
#This test indicates that there is a statistically significant relationship between Indian GDP and manufacturing output, measured in billions of US dollars
#The correlation is weak to moderate from the correlation coefficient of 0.32
#plotted data could not be fitted properly to the linear regression model






```

```python
sns.lineplot(x='year', y='GDP', data=India, label='GDP')

plt.title('India GDP Over Time')
plt.xlabel('Year')
plt.ylabel('Value')

plt.show()
```

```python
sns.lineplot(x='year', y='billions_of_us', data=India, label='Manufacturing Output')


plt.title('India Manufacturing Output Over Time')
plt.xlabel('Year')
plt.ylabel('Value')
```
