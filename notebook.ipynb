| Column | Description |

| `description` | Description of the show |
| `genre` | Show genre |

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")

#finding movie type
Movies_sub = netflix_df[netflix_df['type']=='Movie']

#finding 90s-00s movies
nineties_2000s = Movies_sub[(Movies_sub['release_year']>=1990) & (Movies_sub['release_year']<2000)]

duration_90s_2000s = nineties_2000s['duration']
plt.hist(duration_90s_2000s, edgecolor = 'black')
plt.show()
duration = 100
```

```python
# The issue might be due to the genre column having inconsistent capitalization or extra spaces.
# Let's standardize the genre column to lowercase and strip any extra spaces before filtering.

# Standardize the genre column
nineties_2000s['genre'] = nineties_2000s['genre'].str.lower().str.strip()

# Filter for action movies
short_action_movies = nineties_2000s[nineties_2000s['genre'] == 'action']

# Initialize the counter
short_movie_count = 0

# Iterate through the DataFrame to count short action movies
for label, row in short_action_movies.iterrows():
    if row['duration'] < 90:
        short_movie_count += 1

# Display the count of short action movies
short_movie_count
```
